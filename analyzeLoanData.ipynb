```python
#Task 2

import random

sample_size = random.randint(750, 1250)
sample_data = loan_data.sample(n=sample_size, random_state=42)

print(f"Sample size: {sample_size}")
print(sample_data.head())

sample_data.to_excel('RandomSample_LoanDataset.xlsx', index=False)
```

```python
#Task3
import scipy.stats as stats


mean_loan_amnt = sample_data['loan_amnt'].mean()
std_loan_amnt = sample_data['loan_amnt'].std()
```

```python
mean_loan_amnt
```

```python
std_loan_amnt
```

```python
n = len(sample_data)
```

```python
n
```

```python
#  Z-score for a 95% confidence interval 
z = stats.norm.ppf(0.975)  
```

```python
z
```

```python
margin_of_error = z * (std_loan_amnt / np.sqrt(n))
confidence_interval = (mean_loan_amnt - margin_of_error, mean_loan_amnt + margin_of_error)
```

```python
print(f"Z-score for a 95% confidence interval : {confidence_interval}")
```

```python
#Visualising


plt.figure(figsize=(10, 6))
sns.histplot(sample_data['loan_amnt'], kde=True, color='skyblue', bins=30)
plt.axvline(mean_loan_amnt, color='blue', linestyle='--', label='Mean Loan Amount')
plt.axvline(confidence_interval[0], color='red', linestyle='--', label='Lower Bound (95% CI)')
plt.axvline(confidence_interval[1], color='green', linestyle='--', label='Upper Bound (95% CI)')
plt.title('Loan Amount Distribution with 95% Confidence Interval')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Task 4

# (Mortgage) and (Own)

mortgage_customers = sample_data[sample_data['home_ownership'] == 'MORTGAGE']['loan_amnt']
own_customers = sample_data[sample_data['home_ownership'] == 'OWN']['loan_amnt']
```

```python
mortgage_customers
```

```python
own_customers
```

```python
t_stat, p_value = stats.ttest_ind(mortgage_customers, own_customers, equal_var=True)


print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
print('   ')
# Hypothesis:
if p_value < 0.05:
    print("Null hypothesis rejected: There is a significant difference between the loan amounts.")
else:
    print("Fail to reject null hypothesis: There is no significant difference between the loan amounts.")
```

```python
#Task 5


grouped_data = [sample_data[sample_data['loan_intent'] == intent]['loan_amnt'] 
                for intent in sample_data['loan_intent'].unique()]

# ANOVA test
f_stat, p_value = stats.f_oneway(*grouped_data)


print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")


if p_value < 0.05:
    print("Null hypothesis rejected: There is a significant difference in loan amounts across loan intents.")
else:
    print("Fail to reject null hypothesis: There is no significant difference in loan amounts across loan intents.")
```

```python
# Task 6

#  (home_ownership nd loan_intent )
contingency_table = pd.crosstab(sample_data['home_ownership'], sample_data['loan_intent'])

# KChi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)


print(f"Chi2 Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print('  ')
print(f"Expected Frequencies: \n{expected}")
print('  ')


if p_value < 0.05:
    print("Null hypothesis rejected: There is a significant relationship between home ownership and loan intent.")
else:
    print("Fail to reject null hypothesis: There is no significant relationship between home ownership and loan intent.")
```
